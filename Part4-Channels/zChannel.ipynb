```python
import numpy as np
import matplotlib.pyplot as plt

p = np.arange(1e-10,1,1e-3)
Hp_X_Y = 0.5*( np.log2(1+p) - p*np.log2(p/(1+p)) )
Ip_XY = 1 - Hp_X_Y
                   
fig = plt.figure(figsize=(6,3))

ax = fig.add_subplot(111)
ax.plot(p,Ip_XY, 'b-', label='$I(X;Y)$')
ax.plot(p,Hp_X_Y, 'r-', label='$H(X|Y)$')

plt.xlabel('$p$')
plt.grid(True)
plt.legend(loc=4)
plt.title('$P_0=0.5$')
              
```

```python
P0 = np.arange(1e-10,1,1e-3)
H_X_Y = -P0 * np.log2(2*P0/(1+P0)) - (1-P0)/2*np.log2((1-P0)/(1+P0)) 
H_X = -P0*np.log2(P0) - (1-P0)*np.log2(1-P0)
I_XY = H_X - H_X_Y
                   
fig2 = plt.figure(figsize=(6,3))

ax2 = fig2.add_subplot(111)
ax2.plot(P0,I_XY, 'b-', label='$I(X;Y)$')
ax2.plot(P0,H_X_Y, 'r-', label='$H(X|Y)$')
ax2.plot(P0,H_X, 'g-', label='$H(X)$')

plt.xlabel('$P_0$')
plt.grid(True)
plt.legend(loc=1)
plt.title('$p=0.5$')
```

```python
# capacity
ptr = np.argmax(I_XY)
print('p=%.2f, C=%.2f' %(p[ptr],I_XY[ptr]))
```
