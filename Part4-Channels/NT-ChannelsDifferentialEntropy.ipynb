# Differentielle Entropie

Für kontinuierliche stochastische Signale werden Wahrscheinlichkeitsdichten anstelle von Wahrscheinlichkeiten verwendet. Damit geht die Summe in der Definition der klassischen Entropie in ein Integral über. Es gilt

\begin{align}
    H({\cal X}) = - \sum_{x \in \mathbb{X}} \Pr \{ {\cal X}=x \} \cdot \log_2 \Pr \{ {\cal X}=x \}
    \qquad \Rightarrow \qquad
    h({\cal X}) = - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 p_{\cal X}(x) \ dx \; .
\end{align}

Im Unterschied zur klassichen Entropie $H({\cal X})$, welche nicht negativ wird, kann die differentielle Entropie auch negative Werte annehmen, wie nachfolgende Beispiele zeigen. Damit ist nicht die gleiche Deutung wie im diskreten Fall möglich. Die Differenz von differentiellen Entropien zur Ermittlung der Transinformation erlaubt aber wieder die klassische Interpretation. Es gilt

\begin{align}
    I({\cal X};{\cal Y}) = h({\cal X}) - h({\cal X} \mid {\cal Y}) \geq 0 \; .
\end{align}



## Differentielle Entropie einer Gleichverteilung 

Für eine Gleichverteilung mit Mittelwert $b$ gilt

\begin{align}
    p_{\cal X}(x) &= 
    \begin{cases}
        \frac{1}{2a} & \text{für } |x-b| < a \\
        0 & \text{sonst} \; .
    \end{cases}
\end{align}

Die Leistung der Gleichverteilung beträgt 

\begin{align}
    S &= \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx
       = \frac{1}{2a} \cdot \int_{-a+b}^{a+b} x^2 dx
       = \frac{1}{2a} \cdot \left. \frac{x^3}{3} \right|_{-a+b}^{a+b}
       = \frac{a^2}{3} + b^2 \; ,
\end{align}
und für die Varianz gilt $\sigma_{\cal X}^2 = S- b^2 = \frac{a^2}{3}$. 

Die differentielle Entropie ergibt sich damit zu

\begin{align}
    h({\cal X}) &= - \int_{-a+b}^{a+b} \frac{1}{2a} \cdot \log_2 \frac{1}{2a} dx
                 = \frac{1}{2a} \cdot \log_2(2a) \cdot 2a
                 = \log_2(2a) \; .
\end{align}

Sie ist unabhängig vom Mittelwert, d.h. $b$ enthält keine Information. Da der Mittelwert aber trotzdem zur Leistung beiträgt, werden in der Nachrichtentechnik in der Regel immer mittelwertfreie Sendesignale verwendet. In dem Fall sind Leistung und Varianz identisch und die differentielle Entropie kann auch durch

\begin{align}
    h({\cal X}) &= \log_2(2a) 
                 = \log_2 \left( 2 \sqrt{3 \sigma_{\cal X}^2} \right)
                 = \frac{1}{2} \cdot \log_2 (12 \sigma_{\cal X}^2)  \; .
\end{align}

ausgedrückt werden.


## Differentielle Entropie einer symmetrischen Dreiecksverteilung

Wir haben bereits gesehen, dass der Mittelwert keinen Einfluss auf die differentielle Entropie hat. Daher betrachten wir im folgenden lediglich mittelwertfreie Prozesse. Für eine symmetrische Dreiecksverteilung gilt

\begin{align}
    p_{\cal X}(x) &= 
    \begin{cases}
        \frac{1}{a} \left( 1 - \frac{|x|}{a} \right) & \text{für } |x| \leq a \\
        0 & \text{sonst} \; .
    \end{cases}
\end{align}

Die Leistung der Dreiecksverteilung beträgt 

\begin{align}
    S &= \sigma_{\cal X}^2
       = \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx
       = \frac{2}{a} \cdot \int_0^a x^2 \left( 1 - \frac{x}{a} \right) dx
       = \frac{2}{a} \cdot \left( \left. \frac{x^3}{3} \right|_0^a -  \left. \frac{x^4}{4a} \right|_0^a \right)
       = \frac{a^2}{6} \; .
\end{align}

Die differentielle Entropie ergibt sich damit zu

\begin{align}
    h({\cal X}) 
    &= - \int_{-a}^a p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{a} \left( 1 - \frac{|x|}{a} \right) \right) dx \\
    &= - \log_2 \frac{1}{a} \cdot \int_{-a}^a p_{\cal X}(x) dx
       - \int_{-a}^a p_{\cal X}(x) \cdot \log_2 \left( 1 - \frac{|x|}{a} \right) dx \\  
    &= \vdots \\
    &= \frac{1}{2} \cdot \log_2(ea^2) = \frac{1}{2} \cdot \log_2(6e\sigma_{\cal X}^2) \; .
\end{align}

## Differentielle Entropie einer Gaußverteilung

Für eine Normalverteilung gilt

\begin{align}
    p_{\cal X}(x) &= \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}} \cdot e^{ - \frac{x^2}{2 \sigma_{\cal X}^2} }
\end{align}

wobei die Leistung gleich der Varianz ist. Die differentielle Entropie ergibt sich zu

\begin{align}
    h({\cal X}) 
    &= - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}}
        \cdot e^{ - \frac{x^2}{2 \sigma_{\cal X}^2} } \right) dx \\
    &= - \log_2 \frac{1}{\sqrt{2 \pi \sigma_{\cal X}^2}} \cdot \int_{-\infty}^{\infty} p_{\cal X}(x) dx
       + \frac{\log_2 e}{2 \sigma_{\cal X}^2} \cdot \int_{-\infty}^{\infty} x^2 \cdot p_{\cal X}(x) dx \\  
    &= \frac{1}{2} \cdot \log_2(2 \pi \sigma_{\cal X}^2) + \frac{1}{2} \log_2 e \\  
    &= \frac{1}{2} \cdot \log_2(2 \pi e \sigma_{\cal X}^2) \; .
\end{align}

## Differentielle Entropie einer Laplace-Verteilung 

Für eine Laplace-Verteilung gilt

\begin{align}
    p_{\cal X}(x) &= \frac{1}{\sqrt{2 \sigma_{\cal X}^2}} \cdot e^{ - \frac{\sqrt{2} |x|}{\sigma_{\cal X}} }
\end{align}

wobei die Leistung gleich der Varianz $\sigma_{\cal X}^2$ ist. Die differentielle Entropie ergibt sich zu

\begin{align}
    h({\cal X}) 
    &= - \int_{-\infty}^{\infty} p_{\cal X}(x) \cdot \log_2 \left( \frac{1}{\sqrt{2 \sigma_{\cal X}^2}}
        \cdot e^{ - \frac{\sqrt{2} |x|}{\sigma_{\cal X}} } \right) dx \\
    &= - \log_2 \frac{1}{\sqrt{2 \sigma_{\cal X}^2}} \cdot \int_{-\infty}^{\infty} p_{\cal X}(x) dx
       + \frac{\sqrt{2}}{\sigma_{\cal X}} \cdot \log_2 e \cdot 2 \int_0^{\infty} x \cdot p_{\cal X}(x) dx \\  
    &= \frac{1}{2} \cdot \log_2(2 \sigma_{\cal X}^2) + \log_2 e \\  
    &= \frac{1}{2} \cdot \log_2(2 e^2 \sigma_{\cal X}^2) \; .
\end{align}

## Vergleich der differentiellen Entropien

Ein Vergleich der obigen Ergebnisse zeigt, dass die differentiellen Entropien für die betrachteten Verteilungen immer logarithmisch von der Varianz abhängen. Sie unterscheiden sich daher nur durch konstante Offsets. Für gleiche Varianzen (Leistungen) besitzt die Gaußverteilung die größte differentielle Entropie. Es kann mathematisch bewiesen werden, dass die Gaußverteilung auch generell die größte differentielle Entropie von allen Verteilungen mit endlicher Leistung hat. 

Die Gleichverteilung besitzt von den vier Verteilungen die kleinste differentielle Entropie. Bei diskreten Verteilungen maximiert sie hingegen die Entropie.

```python
import numpy as np
import matplotlib.pyplot as plt

# variances
sigmaX2_log = np.linspace(-20.0,20.0,100)
sigmaX2 = 10**(sigmaX2_log/10.0)

# differential entropy for uniform distribution
h_uniform = 0.5 * np.log2(12*sigmaX2)

# differential entropy for triangular distribution
h_triangular = 0.5 * np.log2(6*np.exp(1)*sigmaX2)

# differential entropy for normal distribution
h_gaussian = 0.5 * np.log2(2*np.pi*np.exp(1)*sigmaX2)

# differential entropy for Laplace distribution
h_laplace = 0.5 * np.log2(2*np.exp(1)**2*sigmaX2)


fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
ax.plot(sigmaX2_log, h_uniform, label='uniform')
ax.plot(sigmaX2_log, h_triangular, label='triangular')
ax.plot(sigmaX2_log, h_gaussian, label='Gaussian')
ax.plot(sigmaX2_log, h_laplace, label='Laplace')
plt.legend()
plt.grid(True)
plt.xlabel('$10 \log_{10} \sigma_{\cal X}^2$')
plt.title('differential entropies')
```

Vergleicht man die erforderlichen Leistungen um die gleiche differentielle Entropie zu erhalten, ergeben sich folgende Unterschiede. Die erste Zeile zeigt die Leistungsvorteile gegenüber einer Gleichverteilung. Die Gaußverteilung hat den größten Vorteil von $\frac{\pi e}{6}=1.42$.

|      -             | Gleichverteilung  | Dreiecksverteilung | Gaußverteilung    | Laplace-Verteilung |
|      -----         |        -----      |        -----       |        -----      |        -----       | 
| Gleichverteilung   |        1          | $\frac{e}{2}=1.36$      | $\frac{\pi e}{6}=1.42$ | $\frac{e^2}{6}=1.23$    |
| Dreiecksverteilung | $\frac{2}{e}=0.74$     |     1              | $\frac{\pi}{3}=1.05$ | $\frac{e}{3}=0.91$                |
| Gaußverteilung     | $\frac{6}{\pi e}=0.70$ | $\frac{3}{\pi}=0.95$  |    1              | $\frac{e}{\pi}=0.87$   | 
| Laplace-Verteilung | $\frac{6}{e^2}=0.81$   | $\frac{3}{e}=1.10$      | $\frac{\pi}{e}=1.16$  |    1               |

## Transinformation für Gleichverteilung

Wir gehen im Folgenden von einem binären Alphabet $\mathbb{D}=\{ -1, +1\}$ mit gleichwahrscheinlichen Symbolen aus. Die zugehörige Entropie lautet $H({\mathcal D}) = 1$bit. Für die Transinformation gilt 

\begin{equation}
    I({\mathcal D};{\mathcal Y}) 
    = H({\mathcal D}) - H({\mathcal D} | {\mathcal Y}) 
    = h({\mathcal Y}) - h({\mathcal Y} | {\mathcal D})
    = h({\mathcal Y}) - h({\mathcal N}) \; . 
\end{equation}

Bei gleichverteiltem Rauschen mit Maximalwert $a$ treten nur Fehler bei $a>1$ auf. Für $a<1$ gilt daher $H({\mathcal D} | {\mathcal Y})=0$ und folglich $I({\mathcal D};{\mathcal Y}) = H({\mathcal D}) = 1$bit. Für $a<1$ ist die differentielle Entropie 

\begin{equation}
    h({\mathcal Y}) = - \int_{- \infty}^{\infty} p(y) \cdot \log_2 p(y) \ dy
\end{equation}

zu berechnen. Die Wahrscheinlichkeitsdichtefunktion 

\begin{equation}
    p(y) = \frac{1}{2} \cdot \left[ p(y+1) + p(y-1) \right]
    = \begin{cases}
        \frac{1}{4a} & \text{für } a-1 < |y| < a+1 \\
        \frac{1}{2a} & \text{für } |y| < a-1 \\
        0            & \text{sonst}
      \end{cases}
\end{equation}

besteht in diesem Fall aus zwei überlagerten Rechtecken. Die differentielle Entropie ergibt sich damit zu

\begin{equation}
    h({\mathcal Y}) = 1 + \frac{1}{a} + \log_2(a) 
\end{equation}

und für die Transinformation gilt

\begin{equation}
    I({\mathcal D};{\mathcal Y}) 
    = 1 + \frac{1}{a} + \log_2(a) - \log_2(2a)
    = \frac{1}{a}
    = \sqrt{\gamma} \; .
\end{equation}

In der obigen Gleichung ist $\gamma = \frac{1}{a^2}$ das Signal-Rausch-Verhältnis.

```python
gamma = np.linspace(0,4,100)
MI = np.where(gamma>1,1,np.sqrt(gamma))

plt.figure(figsize=(9,4))
plt.plot(gamma,MI,'b-')
plt.grid()    
#plt.ylim(-20,30)
plt.xlabel('$\gamma$') 
plt.ylabel('$I(D;Y)$')
```
