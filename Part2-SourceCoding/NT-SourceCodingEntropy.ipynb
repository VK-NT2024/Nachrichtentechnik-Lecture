# Beispiele für Entropien von Zufallsprozessen

Zufallsprozess hat Ereignismenge $\mathbb{X} = \{ X_0,X_1,X_2,X_3,X_4 \}$. Die Ereignisse treten mit den Wahrscheinlichkeiten $\Pr \{ X_{\mu} \}$ auf.

```python
import numpy as np

# Definition der Wahrscheinlichkeiten
Pr = np.array([0.3, 0.2, 0.2, 0.15, 0.15])

# Berechnung der symbolweisen Information
H_Xi = - np.log2(Pr)
print(H_Xi,"\n")

# Berechnung der Entropie
H_X = np.inner(Pr,H_Xi)
print("Die Entropie des Zufallsprozesses beträngt %g bit.\n" % (H_X))
```

# Binäre Entropiefunktion

Die binäre Entropiefunktion ist zu $H_2(p) = -p \cdot \log_2(p) - (1-p) \cdot \log_2(1-p)$ definiert.

```python
%matplotlib inline
import sympy as sp
sp.init_printing()

p = sp.symbols('p', real=True)
H_2 = (-p*sp.log(p)-(1-p)*sp.log(1-p))/sp.log(2)
sp.plot(H_2, (p, 1e-10, 1-1e-10), ylabel='$H_2(p)$',axis_center=(0,0));
```
