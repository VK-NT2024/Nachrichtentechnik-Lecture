# Bayes-Optimale Detektion und Fehlerwahrscheinlichkeiten linearer Modulationsverfahren

## Binäre Modulationsverfahren

Wir betrachten zunächst binäre Modulationsverfahren mit $d \in \DD = \{D_0, D_1\}$ und den A-Priori-Wahrscheinlichkeiten $\Pr\{D_0\} = P_0$ und $\Pr\{D_1\} = 1-P_0$. Ohne Beschränkung der Allgemeinheit wird $\DD \subset \RR$ und $D_1>D_0$ angenommen. Bei Übertragung über einen AWGN-Kanal hat das Empfangssymbol die Form

\begin{equation}
    y = d + n \ ,
\end{equation}

wobei $n$ komplexes, weißes gaußverteiltes Rauschen darstellt. Der MAP-Detektor bestimmt nun das Symbol $\hat{d} \in \DD$, welches die größte A-Posteriori-Wahrscheinlichkeit $\Pr \{ d | y \}$ hat. Es gilt 

\begin{align}
    \hat{d}^{\mathsf{MAP}} 
    &= \arg \! \max_{\tilde{d} \in \mathbb{D}} \Pr \{ \tilde{d} | y \}
     = \arg \! \max_{\tilde{d} \in \mathbb{D}} \big[ p(y | \tilde{d}) \cdot \Pr\{ \tilde{d} \} \big] \\
    &= \arg \! \max_{\tilde{d} \in \mathbb{D}} \big[ \log p(y | \tilde{d}) + \log \Pr\{ \tilde{d} \} \big] 
     = \arg \! \max_{\tilde{d} \in \mathbb{D}} \left[ -\frac{\left( y - \tilde{d} \right)^2}{2 \sigma_{\cal N'}^2} + \log \Pr\{ \tilde{d} \} \right] \; .
\end{align}

Da das Sendealphabet als reellwertig angenommen wurde, spielt der Imaginärteil des Rauschens keine Rolle. Daher reicht es, bei der Likelihood-Funktion $p(y | \tilde{d})$ nur die Rauschleistung $\sigma_{\cal N'}^2 = \sigmaN/2$ des Realteils $n'$ zu berücksichtigen. Bei lediglich 2 Hypothesen kann die Maximierung durch eine einfache Schwellwertentscheidung durchgeführt werden. Die optimale Schwelle $\theta$ liegt im Schnittpunkt der Funktionen $p(y | D_0) \cdot \Pr\{ D_0 \}$ und $p(y | D_1) \cdot \Pr\{ D_1 \}$. Es gilt

\begin{align*}
    p(\theta | D_0) \cdot P_0 &= p(\theta | D_1) \cdot ( 1 - P_0 ) \\
    -\frac{\left( \theta - D_0 \right)^2}{\sigmaN} + \log P_0 &= -\frac{ \left(\theta - D_1 \right)^2}{\sigmaN} + \log (1-P_0) \\
    \log \frac{P_0}{1-P_0} &= \frac{ \left(\theta - D_0 \right)^2 - \left(\theta - D_1 \right)^2}{\sigmaN} 
    = \frac{2 \theta (D_1- D_0) - (D_1 - D_0)(D_0 + D_1)}{\sigmaN} \\
    \theta^{\mathrm{opt}} &= \frac{D_0+D_1}{2} + \frac{\sigmaN}{2(D_1-D_0)} \cdot \log \frac{P_0}{1-P_0}
\end{align*}

Für gleichwahrscheinliche Symbole mit $P_0=\frac{1}{2}$ wird der zweite Term zu Null und die optimale Schwelle liegt mittig zwischen den Symbolen $D_0$ und $D_1$. Für eine ungleiche Verteilung der beiden Symbole verschiebt sich $\theta$ in Richtung des weniger wahrscheinlichen Symbols, da der Faktor vor dem Logarithmus nicht negativ ist.

Mit der optimalen Schwelle $\theta^{\mathrm{opt}}$ können nun die symbolspezifischen Fehlerwahrscheinlichkeiten sowie die mittlere Fehlerwahrscheinlichkeit berechnet werden. Es gilt

\begin{align*}
    Q_0\left(\theta^{\mathrm{opt}}\right) 
    &= \int \limits_{\theta^{\mathrm{opt}}}^{\infty} p(y | D_0) \ dy 
     = \frac{1}{\sqrt{\pi}} \cdot \int \limits_{\left(\theta^{\mathrm{opt}}-D_0\right)/\sigmaN}^{\infty} e^{-t^2} \ dt 
     = \begin{cases}
         \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \frac{\left(\theta^{\mathrm{opt}}-D_0\right)^2}{\sigmaN} } \right) & \text{für } \theta^{\mathrm{opt}} > D_0 \\
         1 - \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \frac{\left(\theta^{\mathrm{opt}}-D_0\right)^2}{\sigmaN} } \right) & \text{für } \theta^{\mathrm{opt}} < D_0 
       \end{cases} \\
    Q_1\left(\theta^{\mathrm{opt}}\right) 
    &= \int \limits_{-\infty}^{\theta^{\mathrm{opt}}} p(y | D_1) \ dy 
     = \frac{1}{\sqrt{\pi}} \cdot \int \limits_{-\infty}^{\left(\theta^{\mathrm{opt}}-D_1\right)/\sigmaN} e^{-t^2} \ dt 
     = \begin{cases}
         \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \frac{\left(\theta^{\mathrm{opt}}-D_1\right)^2}{\sigmaN} } \right) & \text{für } \theta^{\mathrm{opt}} < D_1 \\
         1 - \frac{1}{2} \cdot \mathrm{erfc} \left( \sqrt{ \frac{\left(\theta^{\mathrm{opt}}-D_1\right)^2}{\sigmaN} } \right) & \text{für } \theta^{\mathrm{opt}} > D_1 
       \end{cases}
\end{align*}

Die Argumente der komplementären Fehlerfunktion können für ein bipolares Alphabet mit $D_0=-D_1=e$ und der Sendeleistung $S=e^2$ zu

\begin{align*}
    \frac{\left(\theta^{\mathrm{opt}}-D_0\right)^2}{\sigmaN}
    &= \frac{\left(\frac{D_0+D_1}{2} + \frac{\sigmaN}{2(D_1-D_0)} \cdot \log \frac{P_0}{1-P_0}-D_0\right)^2}{\sigmaN} 
     = \frac{\left(e + \frac{\sigmaN}{4e} \cdot \log \frac{P_0}{1-P_0}\right)^2}{\sigmaN} 
     = \frac{S}{\sigmaN} \cdot \left(1 + \frac{\sigmaN}{4S} \cdot \log \frac{P_0}{1-P_0}\right)^2 \\
    \frac{\left(\theta^{\mathrm{opt}}-D_1\right)^2}{\sigmaN}
    &= \frac{\left(\frac{D_0+D_1}{2} + \frac{\sigmaN}{2(D_1-D_0)} \cdot \log \frac{P_0}{1-P_0}-D_1\right)^2}{\sigmaN} 
     = \frac{\left(-e + \frac{\sigmaN}{4e} \cdot \log \frac{P_0}{1-P_0}\right)^2}{\sigmaN} 
     = \frac{S}{\sigmaN} \cdot \left(1 - \frac{\sigmaN}{4S} \cdot \log \frac{P_0}{1-P_0}\right)^2
\end{align*}

vereinfacht werden. Damit erhalten wir die mittlere Fehlerwahrscheinlichkeit

\begin{align*}
    \Ps &= P_0 \cdot Q_0(\theta^{\mathrm{opt}}) + (1-P_0) \cdot Q_1(\theta^{\mathrm{opt}})
\end{align*}



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special as sp


# 2-ASK (BPSK) modulation scheme
M = 2
m = int(np.log2(M))

# a-priori probabilities
P0 = np.arange(0.1,0.6,0.1)

# signal-to-noise ratio and noise variance
SNRdB = np.linspace(-10,10,21)
SNR = 10**(SNRdB/10)

# erfc arguments for D0 and D1
logP = np.log(P0/(1-P0))
SNRm, logPm = np.meshgrid(4*SNR,logP)
x = logPm / (4*SNRm)
arg0 = np.sqrt(SNR * ( 1 + x )**2)
arg1 = np.sqrt(SNR * ( 1 - x )**2)

# analytical result for 2-ASK (BPSK)
Ps0 = np.where(x>-1,0.5*sp.erfc(arg0),1-0.5*sp.erfc(arg0))
Ps1 = np.where(x<1,0.5*sp.erfc(arg1),1-0.5*sp.erfc(arg1))
Ps = np.tile(P0.reshape((P0.shape[0],1)),(1,SNR.shape[0])) * Ps0 + np.tile((1-P0).reshape((P0.shape[0],1)),(1,SNR.shape[0])) * Ps1

fig1 = plt.figure(figsize=(7,4))
ax1 = fig1.add_subplot(111)
for run in range(P0.shape[0]):
    ax1.semilogy(SNRdB, Ps[run,:], '-', label='$P_0=$%.1f' %(P0[run]))
plt.ylim(1e-3,1)
plt.xlabel('$10 \cdot \log_{10}(E_{\mathrm{s}}/N_0)$')
plt.ylabel('$P_s$')
plt.grid(True)
plt.legend()
```

Offensichtlich kann die Fehlerrate $\Ps$ durch die Ausnutzung der Sendestatistik verbessert werden. Ohne Verschiebung der Schwelle würden Fehlerraten wie für $P_0=0.5$ erzielt. Allerdings fällt der Gewinn in diesem Fall sehr moderat aus. Für hohe SNR tendiert er gegen Null, für kleine SNR sind die Gewinne größer, die Fehlerraten liegen hier aber trotzdem bei sehr hohen Werten. Weiterhin ist zu beachten, dass für $P_0 \neq 0.5$ die Informationsrate sinkt.

```python
P0 = np.arange(0.001,0.501,0.001)

# signal-to-noise ratio and noise variance
SNRdB = np.linspace(-10,10,3)
SNR = 10**(SNRdB/10)

# erfc arguments for D0 and D1
logP = np.log(P0/(1-P0))
SNRm, logPm = np.meshgrid(4*SNR,logP)
x = logPm / (4*SNRm)
arg0 = np.sqrt(SNR * ( 1 + x )**2)
arg1 = np.sqrt(SNR * ( 1 - x )**2)

# analytical result for 2-ASK (BPSK)
Ps0 = np.where(x>-1,0.5*sp.erfc(arg0),1-0.5*sp.erfc(arg0))
Ps1 = np.where(x<1,0.5*sp.erfc(arg1),1-0.5*sp.erfc(arg1))
Ps = np.tile(P0.reshape((P0.shape[0],1)),(1,SNR.shape[0])) * Ps0 + np.tile((1-P0).reshape((P0.shape[0],1)),(1,SNR.shape[0])) * Ps1

# entropy of source signal
H = - P0*np.log2(P0) - (1-P0)*np.log2(1-P0)

fig2 = plt.figure(figsize=(7,4))
ax2 = fig2.add_subplot(111)
#for run in range(SNR.shape[0]):
#    ax2.semilogy(H, Ps[:,run], '-', label='$SNR=$%d' %(SNRdB[run]))
ax2.semilogy(H, Ps[:,0], 'b-', label='$SNR=$%d' %(SNRdB[0]))
ax2.semilogy(H, Ps[:,1], 'g-', label='$SNR=$%d' %(SNRdB[1]))
#ax2.semilogy(H, Ps[:,2], 'r-', label='$SNR=$%d' %(SNRdB[2]))
#plt.ylim(1e-3,1)
plt.xlabel('$H$')
plt.grid(True)
plt.legend()
```

Das obige Bild zeigt die Fehlerraten über der Entropie $H(\Drv)$ des Sendesignals. Interessanterweise sinkt die Fehlerrate signifikant, wenn die Entropie gegen den Wert Null strebt. In diesem Fall wird fast keine Information übertragen, da fast ausschließlich nur eines der beiden möglichen Symbole übertragen wird. Dieses nun sehr zuverlässige Vorwissen überwiegt die Zuverlässigkeit des Kanals, so dass ein SNR-Unterschied von 10 dB mehr als kompensiert werden kann. Bei einem SNR von +10 dB liegen die Fehlerraten allerdings deutlich darunter.

## Fehlerwahrscheinlichkeit bei MAP-Detektion und verschiedenen A-Priori-Verteilungen

Im obigen Diagramm ist zu erkennen, dass die QAM das effizienteste Verfahren ist. Mit ihr kann für $M=16$ eine Fehlerwahrscheinlichkeit $P_{\mathrm{s}}=10^{-5}$ für ein SNR von 20 dB
erzielt werden. Die 16-PSK benötigt ca. 4 dB mehr Sendeleistung (Faktor 2.5), die 16-ASK sogar 9 dB (Faktor 8). Dies liegt an der effizienteren Verteilung der Symbole in der komplexen Ebene, wodurch bei der QAM bei gleicher Sendeleistung eine größere mittlere euklidische Distanz zwischen den Symbolen erzielt werden kann.

Die PSK hat trotzdem Ihre Daseinsberechtigung, da ihre Symbole auf einem Kreis liegen, also einen konstanten Betrag aufweisen. Dies ist dann von Vorteil, wenn starke Schwankungen des Betrags der komplexen Einhüllenden vermieden werden müssen. Diese können bei schlechten Verstärkern mit kleinem linearen Bereich zu nichtlinearen Verzerrungen führen.
