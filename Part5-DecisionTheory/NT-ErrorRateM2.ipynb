```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf, erfc

EsN0dB = np.arange(20)
EsN0 = 10**(EsN0dB/10)

Pb_bipolar = 0.5 * erfc(np.sqrt(EsN0))
Pb_unipolar = 0.5 * erfc(np.sqrt(EsN0/2))

fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(121)
ax.semilogy(EsN0dB, Pb_bipolar, 'b-',label='bipolar')
ax.semilogy(EsN0dB, Pb_unipolar, 'r-',label='unipolar')
plt.xlim(0,12)
plt.ylim(1e-5,1)
plt.xlabel('$Es/N0 in dB$')
plt.grid(True)
plt.legend()

name = 'part5_ErrorRatesM2.ber'
# np.savez(name,EsN0dB=EsN0dB,Pb_bipolar=Pb_bipolar,Pb_unipolar=Pb_unipolar)
erg = np.stack((EsN0dB, Pb_bipolar,Pb_unipolar), axis=-1)
np.savetxt(name,erg)
```
